```python
"""
Список пассажиров Титаника
Данный датасет взят с открытого источника Github (https://github.com/awesomedata/awesome-public-datasets).
Это имитационные данные
Описание атрибутов датасета.
В выбранном датасете представлены следующие атрибуты:
	PassengerId - Id пассажира (тип Int)
	Survived - показывает, выжил пассажир или погиб (1 - выжил, 0 -погиб) (тип Int)
	Pclass - Номер класса пассажира (тип Int)
	Name - Имя пассажира (тип String)
	Sex - Пол пассажира (тип String)
	Age - Возраст пассажира (тип Int)
	SibSp - Число братьев, сестер или супругов на борту у пассажира (тип Int)
	Parch - Количество родителей или детей каждого пассажира (тип Int)
	Ticket - Номер билета (тип String)
	Fare - Цена билета (тип Float)
	Cabin - номер кабины (тип Float)
	Embarked - Порт, в котором произошла посадка на борт Титаника (тип String)
В данной лабораторной работе будут использоваться атрибуты числового типа
"""


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('titanic.csv')

#Среднее значения для каждого числового атрибута
print("Среднее значение для каждого числового атрибута")
df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].mean()
```

```python
#СКО для числовых атрибутов.
print("СКО для каждого числового атрибута")
df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].std()
```

```python
# Построим гистограмму распределения атрибута Age
hist_color = 'purple'
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=50, color=hist_color, edgecolor='black')
plt.grid()
plt.title("Age distribution")

# Определяем границы выбросов
mean_age = df['Age'].mean()
std_age = df['Age'].std()
outlier_threshold = mean_age + 3 * std_age

# Добавляем линию для обозначения границ выбросов
plt.axvline(outlier_threshold, color='red', linestyle='dashed', linewidth=2, label='Граница выбросов')
plt.legend()
plt.show()

# Определяем, есть ли выбросы
outliers = df[df['Age'] > outlier_threshold]
if len(outliers) > 0:
    print(f"Есть выбросы: {len(outliers)} записей.")
else:
    print("Выбросов нет.")
```

```python
# Построим гистограмму распределения атрибута Survived
hist_color = 'purple'
plt.figure(figsize=(5, 5))
plt.hist(df['Survived'], bins=10, color=hist_color, edgecolor='black')
plt.grid()
plt.title("Survived distribution")
#выбросов нет
```

```python
# Построение гистограммы распределения атрибута Pclass
hist_color = 'purple'
plt.figure(figsize=(5, 5))
plt.hist(df['Pclass'], bins=10, color=hist_color, edgecolor='black')
plt.grid()
plt.title("Pclass distribution")
plt.show()
```

```python
# Построение гистограммы распределения атрибута SibSp
hist_color = 'purple'
plt.figure(figsize=(5, 5))
plt.hist(df['SibSp'], bins=15, color=hist_color, edgecolor='black')
plt.grid(True)
plt.title("SibSp distribution")

# Определение границы для выбросов
mean_sibsp = df['SibSp'].mean()
std_sibsp = df['SibSp'].std()
outlier_threshold = mean_sibsp + 3 * std_sibsp

# Добавление вертикальной линии для обозначения границы выбросов
plt.axvline(outlier_threshold, color='red', linestyle='dashed', linewidth=2, label='Граница выбросов')
plt.legend()
plt.show()

# Определение наличия выбросов
outliers = df[df['SibSp'] > outlier_threshold]
if len(outliers) > 0:
    print("Есть выбросы.")
else:
    print("Выбросов нет.")
```

```python
# Построение гистограммы распределения атрибута Parch
hist_color = 'purple'
plt.figure(figsize=(5, 5))
plt.hist(df['Parch'], bins=10, color=hist_color, edgecolor='black')
plt.grid(True)
plt.title("Parch distribution")

# Определение границы для выбросов
mean_parch = df['Parch'].mean()
std_parch = df['Parch'].std()
outlier_threshold = mean_parch + 3 * std_parch

# Добавление вертикальной линии для обозначения границы выбросов
plt.axvline(outlier_threshold, color='red', linestyle='dashed', linewidth=2, label='Граница выбросов')
plt.legend()
plt.show()

# Определение наличия выбросов
outliers = df[df['Parch'] > outlier_threshold]
if len(outliers) > 0:
    print(f"Есть выбросы")
else:
    print("Выбросов нет.")
```

```python
# Построение гистограммы распределения атрибута Fare
hist_color = 'purple'
plt.figure(figsize=(10, 5))
plt.hist(df['Fare'], bins=50, color=hist_color, edgecolor='black')
plt.grid(True)
plt.title("Fare distribution")

# Определение границы для выбросов (например, 3 сигмы от среднего)
mean_fare = df['Fare'].mean()
std_fare = df['Fare'].std()
outlier_threshold = mean_fare + 3 * std_fare
# Определение наличия выбросов
outliers = df[df['Fare'] > outlier_threshold]
if len(outliers) > 0:
    print(f"Есть выбросы")
else:
    print("Выбросов нет.")
```

```python
# Найдем пропущенные значения для числовых атрибутов
missing_data = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].isnull()
missing_counts = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].isnull().sum()

# Выведем количество пропущенных значений
print("Количество пропущенных значений:")
print(missing_counts)

# Выведем информацию о наличии/отсутствии пропущенных значений
if missing_counts.any():
    print("Пропущенные значения есть у следующих атрибутов:")
    attributes_with_missing_values = missing_counts[missing_counts > 0].index
    for attribute in attributes_with_missing_values:
        print(f"{attribute}: {missing_counts[attribute]} пропущенных значений.")
else:
    print("Пропущенных значений нет.")

"""
Варианты обработки пропущенных значений:
1)Если в набор содержит очень большое количество данных, то можно не учитываеть пропущенные значения.
2)Можно заполнить пропущенные значения нулями
3)Можно заменить пропущенные значения выборочным средним
""" 
```

```python
#Определим корреляцию атрибутов
df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()

"""Положительная корреляция - увеличение одной переменной влечет увеличение другой (корреляция стремится к 1)
Отрицательная корреляция - увеличение одной переменной влечет уменьшение другой (корреляция стремится к -1)
Корреляции нет - переменные независимы (корреляция равна 0)
Наибольшая корреляция у Fare и Pclass: -0.549500.
Корреляцию 0 в нашем случае не имеет ни один атрибут."""
```

```python
# Матрица графиков рассеивания
point_color = 'purple'
variables = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
g = sns.pairplot(df[variables], plot_kws={'color': point_color})
# Настраиваем оси X и Y для каждого графика
for i in range(len(variables)):
    for j in range(len(variables)):
        if i != j:
            g.axes[i, j].xaxis.set_major_locator(plt.MaxNLocator(integer=True))
            g.axes[i, j].yaxis.set_major_locator(plt.MaxNLocator(integer=True))
g.fig.suptitle("Матрица графиков рассеивания", y=1.02)
plt.show()
```
